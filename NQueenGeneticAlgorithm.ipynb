```python
import random
```

```python
def generating_random_queen_rows(n):
  queen_placed = random.sample(range(1, n+1), n)
  return queen_placed
```

```python
def fitness(board, maximum_fitness):
  board_length = len(board)
  attacks = 0
  for i in range(board_length):
    for j in range(i +1, board_length):
      if board[i]==board[j] or abs(i-j)==abs(board[i]-board[j]):
        attacks +=1
  fitness = maximum_fitness -attacks
  return fitness
```

```python
def initial_population(population, n):
  return [generating_random_queen_rows(n) for _ in range(population)]
```

```python
def population_fitness(population):
  return [fitness(i, maximum_fitness) for i in population]
```

```python
def selectionfunct(population, calculating_fitness):
  total_fitness = sum(calculating_fitness)
  prob = [fit_score/total_fitness for fit_score in calculating_fitness]
  selection = random.choices(range(len(population)), weights=prob, k=len(population))

  select_nextgeneration = [population[i] for i in selection]
  return select_nextgeneration
```

```python
def single_point_crossover(parent1, parent2):
  cross_point = random.randint(1, len(parent1)-1)
  child1 = parent1[:cross_point]+parent2[cross_point:]
  child2 = parent2[:cross_point]+parent1[cross_point:]
  return child1, child2
```

```python
def crossover(population):
  random.shuffle(population)

  if len(population)%2 != 0:
    population.pop()
  childs = []
  for i in range(0, len(population), 2):
    parent1, parent2 = population[i] , population[i+1]
    child1, child2 =  single_point_crossover(parent1, parent2)
    childs.extend([child1, child2])

  return childs
```

```python
def mutation(candidate, muatation_rate):
  if random.uniform(0,1) < muatation_rate:
    mutate_position = random.randint(0, len(candidate) -1)
    new_position = random.randint(1, len(candidate))
    candidate[mutate_position] == new_position

  return candidate
```

```python
def applying_mutation(population, mutation_rate):
  mutate_population = [mutation(candidate, mutation_rate) for candidate in population]
  return mutate_population
```
